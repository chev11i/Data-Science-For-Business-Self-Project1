# TASK #6: PERFORM DATA CLEANING BY APPLYING EVERYTHING WE LEARNED SO FAR!

```python
# Let's define a pipeline to clean up all the messages 
# The pipeline performs the following: (1) remove punctuation, (2) remove stopwords
def message_cleaning(message):
  message_punc_removed = [ char for char in message if char not in string.punctuation]
  message_punc_removed_join = ''.join(message_punc_removed)
  message_punc_removed_join_clean = [ word for word in message_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
  return message_punc_removed_join_clean
```

```python
# Let's test the newly added function
reviews_df_clean = reviews_df['verified_reviews'].apply(message_cleaning)
```

```python
print(reviews_df_clean[3])
```

```python
print(reviews_df['verified_reviews'][3])
```

```python
vectorizer = CountVectorizer(analyzer= message_cleaning)
reviews_countvectorizer = vectorizer.fit_transform(reviews_df['verified_reviews'])
```

```python
print(vectorizer.get_feature_names())
```

```python
print(reviews_countvectorizer.toarray())
```

```python
reviews_countvectorizer.shape
```

```python
reviews_df.head()
```

```python
reviews_df_dropped = reviews_df.drop(['verified_reviews','feedback'],axis=1)
reviews = pd.DataFrame(reviews_countvectorizer.toarray())
X = pd.concat([reviews_df_dropped,reviews],axis=1)
y = reviews_df['feedback']
```

# TASK #7: TRAIN A NAIVE BAYES CLASSIFIER MODEL

```python
from sklearn.model_selection import  train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train,y_train)
```

# TASK #8: ASSESS TRAINED MODEL PERFORMANCE  

```python
from sklearn.metrics import classification_report,confusion_matrix

y_predict_train = NB_classifier.predict(X_train)
y_predict_train
```

```python
cm = confusion_matrix(y_train,y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
y_predict_test = NB_classifier.predict(X_test)
y_predict_test
```

```python
cm = confusion_matrix(y_test,y_predict_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predict_test))
```

# TASK #9: ASSIGNMENT - TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.linear_model import LogisticRegression

LR_classifier = LogisticRegression()
LR_classifier.fit(X_train,y_train)

y_pred = LR_classifier.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot = True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy {} %'.format(100* accuracy_score(y_test,y_pred)))
```

```python
print(classification_report(y_test,y_pred))
```
